`WAP to get the largest and smallest number in a list without using built-in function`

```python
list = [6,9,3,5,55,33,76,99,12]
largest = list[0]
smallest = list[0]
for num in list:
    if largest < num:
        largest = num
    if smallest > num:
        smallest = num
print(f"Largest is {largest} and smallest is {smallest}.")
```

`WAP to check if each number is prime in a given list of numbers. Print the prime numbers if any, otherwise print "no prime numbers"`

```python
list = [6,9,3,5,55,33,19,76,99,12]
list1 = []
for num in list:
    count = 0
    for i in range(1,num+1):
        if num  % i == 0:
            count += 1
    if count == 2:
            list1.append(num)
if not list1:
    print("No prime number")
else:
    print(list1)
del list
```

`WAP to insert n values in a list and find those values in a list that are greater than a specified number`

```python
list1 = []
list2 = []
n = int(input())
s = int(input())
for i in range(0,n+1):
    x = int(input())
    list1.append(x)
for i in list1:
    if i > s:
        list2.append(i)
print(list2)
```
