`WAP to convert a  tuple to a string`

`WAP to get the 4th element from the last element of a tuple`

```python
tuplee = (1,2,3,4,5,6,7,8)
print(tuplee[-4])
```

`WAP to find repeated items in a tuple`

```python
tuplee = (1,1,2,3,4,6,6,7,8,9,9)
for num in tuplee:
    if tuplee.count(num) > 1:
        print(num)
```

`WAP to replace the last value of the tuples in a list`

```python
listt = [(10,20,40),(40,50,60),(70,80,90)]

new_listt = [t[:-1] + (100,) for t in listt]
print(new_listt)
```

`WAP to sort a tuple by its float element`

```python

data = (('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5'))
sorted_data = tuple(sorted(data, key=lambda x: float(x[1])))

print("Original:", data)
print("Sorted :", sorted_data)
```

`WAP to convert a given string list to a tuple'

```python
data = 'python3.0'
print(type(data))
new_data = [x for x in data]
new_data1 = tuple(new_data)
print(new_data1)
print(type(new_data1))
```

`WAP to calculate the product multiplying all the numbers in a given tuple`

```python
data = (1,2,3,4,5,6,7,8)
mul = 1
for num in data:
    mul *= num
print(mul)
```
