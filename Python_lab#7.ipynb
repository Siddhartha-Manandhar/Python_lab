# 3
`Recursion`

```python
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
print(fact(5))
```

`Without recursion`

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else: 
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
print(factorial(5))
```

```python
def rectangle_area(length, breadth):
    area = length * breadth
    perimeter = 2 * (length + breadth)
    return area, perimeter
result = rectangle_area(5, 3)
print("Area:", result[0])
print("Perimeter:", result[1])
```

# 6

```python
import math
def circle_area(radius):
    area = math.pi * radius * radius
    circumference = 2 * math.pi * radius
    return area, circumference
result = circle_area(5)
print("Area:", result[0])
print("Circumference:", result[1])
```

# 8

```python
import math
def maxmin(lst):
    return max(lst), min(lst)

def summult(lst):
    total = sum(lst)
    product = math.prod(lst)
    return total, product
numbers = [1, 2, 3, 4, 5]
max_value, min_value = maxmin(numbers)
sum_value, product_value = summult(numbers)
print("Max:", max_value)
print("Min:", min_value)
print("Sum:", sum_value)
print("Product:", product_value)
```

# 12

```python
def calculate_avg(lst):
    if not lst:
        return 0
    return sum(lst) / len(lst)
numbers = [10, 20, 30, 40, 50]
average = calculate_avg(numbers)
print("Average:", average)
```

# 13

```python
def palindrome(s):
    s = s.replace(" ", "").lower()
    return s == s[::-1]
print(palindrome("madam"))
```

# 14

```python
def upper_lower(s):
    upper_count = sum(1 for char in s if char.isupper())
    lower_count = sum(1 for char in s if char.islower())
    return upper_count, lower_count
result = upper_lower("Hello World")
print("Uppercase:", result[0])
print("Lowercase:", result[1])
```

# 15

```python
n = 5  # global variable

def fact():
    global n
    if n == 0 or n == 1:
        return 1
    else:
        n -= 1
        return (n + 1) * fact()

print(fact())
```
