```python
numbers = [1,2,3,4,7,8,9,45,34,23,44,11,56]
def checkeven(numbers):
    even = []
    odd = []
    for i in numbers:
        if i % 2 == 0:
            even.append(i)
        else:
            odd.append(i)
    print("No. of even ", len(even))
    print(even)
    print("No. of odd ", len(odd))
    print(odd)

checkeven(numbers)
```

```python
name = "Team Chaos"
def common(name):
    max_count = 0
    char = ""
    for ch in name:
        count = name.count(ch)
        if count > max_count:
            max_count = count
            char = ch
    print(char, " = ", max_count)
common(name)
```
